Defining High Value Customer subset


*   Phone Service = Yes
*   Internet Service = Yes
*   Total Revenue > $3000 (Greater than the mean)



```python
def df_high_value(row):
  if row["Phone Service"] == "Yes" and row["Internet Service"] == "Yes" and row["Total Revenue"]>= 3000:
    return True
  else:
    return False
```

```python
df_org["High Value Customer"] = df_org.apply(df_high_value, axis=1)
```

```python
df_org["High Value Customer"].value_counts()
```

```python
df_org.groupby("High Value Customer")["Customer Status"].value_counts(normalize= True)
```

Offer acceptance influences retention or not?

```python
df_org["Offer"].unique()
```

```python
def offer_acceptance(row):
  if pd.isna(row["Offer"]):
    return False
  else:
    return True
```

```python
df_org["Offer Acceptance"]= df_org.apply(offer_acceptance, axis=1)
```

```python
df_org["Offer Acceptance"].value_counts()
```

```python
df_org.groupby("Offer Acceptance")["Customer Status"].value_counts(normalize= True)
```
